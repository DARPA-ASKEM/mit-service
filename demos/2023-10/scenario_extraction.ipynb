```python
from langchain.chat_models import ChatOpenAI
from langchain.schema.messages import HumanMessage, SystemMessage

chat_gpt = ChatOpenAI(model_name='gpt-3.5-turbo', openai_api_key=os.environ['OPENAI_API_KEY'], temperature=0, max_tokens=300)
```

```python
import pandas as pd
```

```python
# remove all newlines and extra spaces from the strings
def clean_string(s):
    return ' '.join(s.split())
```

```python
def flatten_paper_data(paper_data):
    flat_data = []
    for paper in paper_data:
        for i, d in enumerate(paper['figure_captions']):
            ## replace empty
            flat_data.append({'name': paper['name'], 'caption': clean_string(d['caption'])})
    return flat_data
```

```python
data = flatten_paper_data(paper_data)
```

```python
df = pd.DataFrame(data)
```

```python
## change the styler for the caption column to show a larger text area and allow wrapping
def caption_styler(caption):
    return f'<textarea rows="20" cols="60">{caption}</textarea>'


def show_df(df):
    return df.style.format({'content': caption_styler})
```

```python
import pandas as pd
cosmos_dir = pathlib.Path('~/Downloads/cosmos-4papers/').expanduser()
cosmos_dfs = []

for folder in cosmos_dir.iterdir():
    if folder.is_dir():
        for file in folder.iterdir():
            if file.suffix == '.parquet':
                if file.stem.replace(' ', '_') == folder.name:
                    cosmos_dfs.append(pd.read_parquet(str(file)))


extraction_df= pd.concat(cosmos_dfs, ignore_index=True)

caption_df = extraction_df[extraction_df.postprocess_cls == 'Figure Caption']
caption_df = caption_df[['pdf_name', 'content']]
```

```python
template = f'''For the following paragraph, return a json list of all geographic locations mentioned therein. If no locations are mentioned, return an empty list.'''

def extract_locations(caption):
    if not caption:
        return []
    return json.loads(chat_gpt.invoke(input=[SystemMessage(content=template), HumanMessage(content=caption)]).content)

caption_df = caption_df.assign(locations=caption_df['content'].map(extract_locations))
show_df(caption_df)
```
